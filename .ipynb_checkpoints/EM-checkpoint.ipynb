```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
```

```python
X, y_true = make_blobs(n_samples=3000, centers=30, cluster_std=0.60, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, -1], s=40)
plt.title("Synthetic Cluster Data")
plt.show()
```

```python
gmm = GaussianMixture(n_components=2, random_state=42)
gmm.fit(X)
labels = gmm.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, -1], c=labels, cmap='viridis', s=40)
plt.title("Clusters formed by GMM")
plt.show()
```

```python
score = silhouette_score(X, labels)
print("Silhouette Score: ", score)
```

```python
print("Means of clusters:\n", gmm.means_)
```

```python
print("Covariances of clusters:\n", gmm.covariances_)
```
