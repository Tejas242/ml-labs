```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
%matplotlib inline
```

```python
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.6, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=40)
plt.title("Generated Data")
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=3, linkage='ward') # 'ward' minimizes variance of merged clusters
labels = model.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=40)
plt.title("Predicted Clusters")
plt.show()
```

```python
Z = linkage(X, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title("Dendrogram for hierarchical Clustering")
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
