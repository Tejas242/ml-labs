```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=df)
plt.title('Target Class Distribution')
plt.show()
```

```python
corr = df.corr()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(corr, cmap='coolwarm')
plt.show()
```

```python
selected_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness', 'target']
sns.pairplot(df[selected_features], hue='target')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
df.drop('target', axis=1).boxplot()
plt.xticks(rotation=90)
plt.title('Boxplot of Features')
plt.show()
```

```python
df.drop('target', axis=1).skew()
```

```python
from scipy.stats import zscore
z_scores = np.abs(zscore(df.drop('target', axis=1)))
outliers = (z_scores > 3).sum(axis=0)
print(outliers)
```

```python
for c in df.columns[:-1]:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[c], kde=True)
    plt.title(f'Distribution of {c}')
    plt.show()
```
